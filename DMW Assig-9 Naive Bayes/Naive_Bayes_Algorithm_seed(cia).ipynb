```python
# Import libraries
import numpy as np
import pandas as pd
```

```python
# Read data from dataset into dataframe 
df = pd.read_csv('seeds.csv')
```

```python
df
```

```python

# Display first 5 rows
df.head(10)
```

```python
df.isnull().sum()
```

```python
df.columns=['Area', 'Perimeter', 'Compactness', 'Kernel_Length','Kernel_Width', 'Asymmetry_Coeff','Kernel_Groove' , 'Category']
df.head()
```

```python
#Data Transformation
df["Category"].replace({1 :"Kama" , 2 :"Rosa" , 3 :"Canadian"} , inplace = True)
df.head()
```

```python
#Data Visualization
import matplotlib.pyplot as plt
df['Category'].value_counts().plot(kind='bar', title='Wheat Seeds Categories', figsize=(15,5))
plt.show()
```

```python
#Correlation matrix
import seaborn as sns
corr = df.corr()
fig, ax = plt.subplots(figsize=(10,10)) 
sns.heatmap(corr, annot=True, square=True ,ax=ax)
plt.yticks(rotation=360)
plt.show()
```

```python
#Plot
sns.relplot(x='Compactness', y='Area', hue='Category', data=df);
```

```python
x = df.iloc[:, 0:7]
x.head()
```

```python
y = df['Category']
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x, y, random_state=50, test_size = 0.25)
```

```python
from sklearn.naive_bayes import GaussianNB

#Create a Gaussian Classifier
model = GaussianNB()

# Train the model using traning sets
model.fit(x_train, y_train)

# Predict the response for test datasets
Y_pred = model.predict(x_test)
```

```python
import sklearn.metrics as metrics
```

```python
print("accuracy:", metrics.accuracy_score (y_test, Y_pred))
```
